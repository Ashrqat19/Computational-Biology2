```python
import networkx as nx
```

```python
A = nx.Graph()
A.add_nodes_from(['a', 'm', 's', 'd','r'])
A.add_edges_from([('a','s'), ('m','d'), ('s','m'), ('a','d'),('r','a'),('d','r')])
nx.draw(A, with_labels = True)
```

```python
list(A.neighbors('a'))
```

```python
for neighbor in A.neighbors('d'):
    print(neighbor)
```

```python
A.has_node('a')
```

```python
A.has_node('f')
```

```python
nx.is_tree(A)
```

```python
nx.is_connected(A)
```

```python
A.degree('m')
```

```python
S = nx.DiGraph()
S.add_edges_from([('a','s'), ('m','d'), ('s','m'), ('a','d')])
nx.draw(S, with_labels = True,node_size = 600)
```

```python
dict(S.degree())
```

```python
nx.to_numpy_matrix(S)
```
